# Polynomials and Splines
This question uses the variables \texttt{\color{brown}dis} (the weighted mean of distances to five Boston employment centers) and \texttt{\color{brown}nox} (nitrogen oxides concen- tration in parts per 10 million) from the \texttt{\color{brown}Boston} data. We will treat \texttt{\color{brown}dis} as the predictor and \texttt{\color{brown}nox} as the response.

$(a)$ Use the \texttt{\color{brown}poly()} function to fit a cubic polynomial regression to predict \texttt{\color{brown}nox} using \texttt{\color{brown}dis}. Report the regression output, and plot the resulting data and polynomial fits.
\newline


First we initialize the data, from the library \texttt{MASS}. Note that here \texttt{nox} is the response, hence the $Y-$axis, and \texttt{dis} is the predictor, hence the $X-$axis. We start by visualizing the data.

```r
library(MASS)
data = Boston
plot(data$nox ~ data$dis,  xlab = "dis", ylab = "nox")
```

Next we fit a model using cubic polynomial regression with the \texttt{poly()} function.

We can see that there's some sort of a local minima ar around $4$, and then a bend at around $8$. This suggests that using quadratic polynomial is not a good idea.

```r
fit <- lm(nox ~ poly(dis, 3), data)
summary(fit)
```

Next, we'll plot the cubic polynomial that we got from the regression.

```r
#getting the points to input into the polynomial function, to draw the graph of polynomial
#these are the 'test' points
dis.test = seq(from = min(data$dis), to = max(data$dis), by = 0.01)

#find response for these 'test' points
nox.test = predict(fit, list(dis = dis.test))

#plot the graph now
plot(data$nox ~ data$dis,  xlab = "dis", ylab = "nox")
lines(dis.test, nox.test, col = "red", lwd = 2)
```

$(b)$ Plot the polynomial fits for a range of different polynomial degrees (say, from 1 to 10), and report the associated residual sum of squares.\newline

To do this, we simply do the above, but for every degree from 1 to 10. We plot all the polynomials (with different colors) in the same graph for comparison. We will use the input as \texttt{dis.test} from above

```r
plot(data$nox ~ data$dis,  xlab="dis", ylab="nox") #make a 'base' plot
deg.lim = 10 #initialize how many degrees we want
color = rainbow(10) #generate colors
degrees = 1:deg.lim #vector to store the list of degrees
MSE = 1:deg.lim #vector to store the mean square error for each degree

#do the same for each degree as we did for degree 3
for (i in degrees) {
    fit = lm(nox ~ poly(dis, i), data) #train the model
    prediction = predict(fit, list(dis = dis.test)) #get the prediction
    lines(dis.test, prediction, col = color[i], lwd = 2) #plot the polynomial
    MSE[i] = mean(fit$residuals^2)
    }
#add a legend for associating colors to degrees
legend(x = 'topright',legend = 1:deg.lim, col = color, lty = c(1,1), lwd = c(2,2), bty = "n", cex = 0.7)
```

```r
#make a dataframe for mean squared errors
polyMSE = data.frame(degrees,MSE)
plot(polyMSE, type="o")
```

$(c)$ Perform cross validation or another approach to select the optimal degree for the polynomial, and explain your results.

```r
library(boot)
cv_mse = 1:deg.lim #initialize array to store cv-errors
for (i in degrees){
    glfit = glm(nox ~ poly(dis, i), data=data) #fit the model
    cv_mse[i] = cv.glm(data, glfit)$delta[1] #do cross validation and find the cv MSE
}
```

```r
plot(cv_mse ~ degrees, type="o")
cat("Minimum occurs at:",which.min(cv_mse))
```

From here we can say that a cubic polynomial minimizes the CV MSE.

$(d)$ Use the \texttt{\color{brown}bs()} function to fit a regression spline to predict “\texttt{\color{brown}nox}” using “\texttt{\color{brown}dis}”. Report the output for the fit using four degrees of freedom. How did you choose the knots? Plot the resulting fit.

```r
#We need the 'splines' library for the bs()
library(splines)
attr(bs(data$dis, df = 4), "knots")
```

The knots were chosen using the \texttt{attr(bs())} function which has R automatically determine the best knots. R chose $1$ knot at $3.20745$.

```r
bs_fit = lm(nox ~ bs(dis, knots = c(3.20745)), data = data)
summary(bs_fit)
```

```r
#now we test this, that is, predict
bs.test = predict(bs_fit, list(dis = dis.test))
plot(nox ~ dis, data = data) 
lines(dis.test, bs.test, col = "red", lwd = 2)
```

$(e)$ Now fit a regression spline for a range of degrees of freedom, and plot the resulting fits and report the resulting RSS. Describe the results obtained.

```r
df.lim = 20 #number of degrees of freedom we want to check
dfs = 3:(2 + df.lim) #array to store the degrees of freedom
color = rainbow(df.lim) #generate colors for each degree of freedom
rss = 1:deg.lim #vector to store the mean square error for each degree
plot(nox ~ dis, data = data) #first plot the 'nox' and 'dis'
#fit a model for each df
for (i in dfs){
    bs_fit = lm(nox ~ bs(dis, df = i), data = data)
    prediction = predict(bs_fit, list(dis = dis.test))
    lines(prediction ~ dis.test, col = color[i])
    rss[i] = mean(bs_fit$residuals^2)
}
#add a legend for associating colors to degrees
legend(x = 'topright',legend = dfs, col = color, lty = c(1,1), lwd = c(2,2), bty = "n", cex = 0.7)
```

$(f)$ Perform cross-validation or another approach in order to select the best degrees of freedom for a regression spline on this data. Describe your results.

```r
bs_mse = 3:(2+df.lim) #initialize array to store cv-errors in spline fitting
for (i in dfs){
    model = model.frame(nox~bs(dis,df = i), data = data) #create the model frame for df=i
    names(model) = c('nox','dis') #give names to columns
    bs_fit = glm(nox ~ dis, data = model) #fit the model
    bs_mse[i-2] = cv.glm(bs_fit, data = model)$delta[1] #find errors
}
plot(dfs, bs_mse,type = 'o', xlab = 'Deg of freedom', ylab = 'Cross Validated MSE for Splines')
```

From here it is visible that our 'best' degree of freedom is $10$.

# Decision Tree
This problem involves the \texttt{\color{brown}OJ} data set which is part of the \texttt{\color{brown}ISLR2} package.

```r
#start by importing the required packages and data
set.seed(122)
library(ISLR2)
library(tree)
library(rpart)
data(OJ)
```

$(a)$ Create a training set containing a random sample of $800$ observations, and a test set containing the remaining observations.

```r
train_row = sample(nrow(OJ), 800) #get training row numbers
oj.train = OJ[train_row,] #get training rows
oj.test = OJ[-train_row,] #get testing rows
```

$(b)$ Fit a tree to the training data, with \texttt{\color{brown}Purchase} as the response and the other variables as predictors. Use the \texttt{\color{brown}summary()} function to produce summary statistics about the tree, and describe the results obtained. What is the training error rate? How many terminal nodes does the tree have?

```r
oj.tree = tree(Purchase ~ . , data = oj.train)
summary(oj.tree)
```

We can see that the number of terminal nodes in the tree is $8$ and the training error is $0.165$. The tree only uses two variables: ``LoyalCH" and ``PriceDiff"

$(c)$ Type in the name of the tree object in order to get a detailed text output. Pick one of the terminal nodes, and interpret the information displayed.

```r
oj.tree
```

Let's pick the terminal node labelled "4)". The $\ast$ against it denotes that it's terminal. The variable whose value is looked for splitting is "LoyalCH", and this splitting value is $0.064156$. This node has $72$ observations in the subtree below this, with deviance $18.28$. The overall prediction is "MM" - about $2.8%$ of the observations have prediction as "CH" and the rest $97.2%$ have "MM".

Let's also interpret the node labelled "25)". Again, $\ast$ denotes that it's terminal. The variable whose value is looked for splitting is "PriceDiff", and this splitting value is $-0.35$. This node has $122$ observations in the subtree below this, with deviance $160.60$. The overall prediction is "CH" - about $87%$ of the observations have prediction as "CH" and the rest $13%$ have "MM".

$(d)$ Create a plot of the tree, and interpret the results.

```r
plot(oj.tree)
text(oj.tree, cex = 0.6)
```

The top $3$ nodes have splitting variable "LoyalCH". Hence this is the most important variable. Note that if $\textrm{LoyalCH}<0.064156$ then the prediction is "MM" and if $\textrm{LoyalCH}>0.764572$ then the prediction is "CH" - this is a wide range of values. The second parameter "PriceDiff" comes into play only when $\textrm{LoyalCH}\in [0.064156,0.764572]$.

$(e)$ Predict the response on the test data, and produce a confusion matrix comparing the test labels to the predicted test labels. What is the test error rate?

```r
oj.prediction =  predict(oj.tree, oj.test, type="class")

cat('Confusion matrix:\n')
print(table(oj.test$Purchase, oj.prediction))
error.test = mean(oj.test$Purchase != oj.prediction)
cat('Testing error rate:', round(error.test*100,2), '%')
```

$(f)$ Apply the  \texttt{\color{brown}cv.tree()} function to the training set in order to determine the optimal tree size.

```r
oj.cvtree = cv.tree(oj.tree) 
```

$(g)$ Produce a plot with tree size on the x-axis and cross-validated classification error rate on the y-axis.

```r
plot(oj.cvtree$dev ~ oj.cvtree$size, type = "o", xlab = "Tree Size", ylab = "Deviance")
print(data.frame("Deviance" = oj.cvtree$dev, "Size" = oj.cvtree$size))
```

$(h)$ Which tree size corresponds to the lowest cross-validated classification error rate?

```r
cat("Size corresponding to minimum deviance by cross-validation:", oj.cvtree$size[which.min(oj.cvtree$dev)])
```

$(i)$ Produce a pruned tree corresponding to the optimal tree size obtained using cross-validation. If cross-validation does not lead to selection of a pruned tree, then create a pruned tree with five terminal nodes.

```r
oj.pruned_tree <-  prune.tree(oj.tree, best = 5)
plot(oj.pruned_tree)
text(oj.pruned_tree)
```

$(j)$ Compare the training error rates between the pruned and unpruned trees. Which is higher?

```r
summary(oj.tree)
summary(oj.pruned_tree)
```

```r
#we report the training error comparison

unprune.train_prediction = predict(oj.tree, oj.train, type = "class")
unpruned.train_error =  mean(oj.train$Purchase != unprune.train_prediction)
cat("Error for unpruned tree:\t", unpruned.train_error, "\n")

pruned.train_prediction = predict(oj.pruned_tree, oj.train, type = "class")
pruned.train_error =  mean(oj.train$Purchase != pruned.train_prediction)
cat("Error for pruned tree:\t\t", pruned.train_error,"\n")

if (unpruned.train_error > pruned.train_error){
    cat("Training error for unpruned tree is more than pruned tree")
} else{
    cat("Training error for pruned tree is more than unpruned tree")
}
```

$(k)$ Compare the testing error rates between the pruned and unpruned trees. Which is higher?

```r
#we report the testing error comparison

unpruned.test_prediction = predict(oj.tree, oj.test, type = "class")
unpruned.test_error =  mean(oj.test$Purchase != unpruned.test_prediction)
cat("Error for unpruned tree:\t", unpruned.test_error, "\n")

pruned.test_prediction = predict(oj.pruned_tree, oj.test, type = "class")
pruned.test_error =  mean(oj.test$Purchase != pruned.test_prediction)
cat("Error for pruned tree:\t\t", pruned.test_error,"\n")

if (unpruned.test_error > pruned.test_error){
    cat("Testing error for unpruned tree is more than pruned tree")
} else{
    cat("Testing error for pruned tree is more than unpruned tree")
}
```
